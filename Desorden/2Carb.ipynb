```python
!hybracter hybrid -h
```

__Requeriments__

Hybracter hybrid require and input .csv file to be specified with --input flag. No other inputs are required.

- This file requires no headers
- Other than the reads, hybracter requires a value for a lower bound the minimum chromosome length for each isolate in base pairs. It must be an integer
- hybracter will denote contigs about this value as chromosome(s) and if it can recover a chromosome, it will denote the isolate as complete
- In practice, I suggest choosing 90% of the estimated chromosome size for this value
- e.g. for S. aureus, I'd choose 2500000, E. coli, 4000000, P. aeruginosa 5500000.

1. **qseqid**: Query sequence ID. Es el identificador de la secuencia de consulta.

2. **sseqid**: Subject sequence ID. Es el identificador de la secuencia objetivo en la base de datos contra la cual se está comparando la secuencia de consulta.
3. **pident**: Percentage of identical matches. Es el porcentaje de identidad entre la secuencia de consulta y la secuencia objetivo.
4. **length**: Alignment length. Es la longitud del alineamiento entre la secuencia de consulta y la secuencia objetivo.
5. **qlen**: Query sequence length. Es la longitud total de la secuencia de consulta.
6. **slen**: Subject sequence length. Es la longitud total de la secuencia objetivo.
7. **evalue**: Expect value. Es un valor que indica el número de alineamientos esperados por casualidad en una base de datos de tamaño determinado. Valores menores indican alineamientos más significativos.
8. **bitscore**: Bit score. Es una medida de la calidad del alineamiento; valores más altos indican mejores alineamientos.
9. **qcov**: Query coverage. Es la cobertura de la secuencia de consulta, que indica qué porcentaje de la secuencia de consulta está alineado con la secuencia objetivo.
10. **sample**: Sample identifier. Es un identificador para la muestra de la que proviene la secuencia de consulta.
11. **scov**: Subject coverage. Es la cobertura de la secuencia objetivo, indicando qué porcentaje de la secuencia objetivo está alineado con la secuencia de consulta.
12. **gene**: Gene name. Es el nombre del gen correspondiente a la secuencia objetivo.
13. **subtype**: Subtype. Es el subtipo de la secuencia objetivo.
14. **type**: Type. Es el tipo de la secuencia objetivo o la clasificación más general.
15. **count**: Count. Es el número de veces que la secuencia de consulta se ha alineado con la secuencia objetivo.
16. **copy**: Copy number. Es el número de copias del gen o secuencia en la muestra de consulta.
17. **rpk**: Reads per kilobase. Es una medida de abundancia que normaliza el número de lecturas por la longitud de la secuencia en kilobases.
18. **rpksum**: Sum of reads per kilobase. Es la suma de todas las lecturas por kilobase para todas las secuencias en la muestra.
